```
try:
    from urllib.request import Request, urlopen  # Python 3
except ImportError:
    from urllib2 import Request, urlopen  # Python 2

req = Request('https://transparencia.registrocivil.org.br/covid/obitos-2020.csv')
req.add_header('User-Agent', 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_1_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36')
content = urlopen(req)

df = pd.read_csv(content)
print(df)
```
